# 01: Explore Satellite Data

Goal: Load Sentinel-2 imagery for a crisis-prone region and visualize NDVI.

```python
# Cell 1: Setup
import sys
sys.path.append('../src')

import numpy as np
import matplotlib.pyplot as plt
from crisiscortex.data.satellite import get_crisis_regions, SentinelDownloader

# Cell 2: Explore regions
regions = get_crisis_regions()
for name, bbox in regions.items():
    print(f"{name}: {bbox}")

# Cell 3: Visualize what NDVI values mean
# Create synthetic NDVI for demonstration
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

# Healthy vegetation
healthy = np.random.uniform(0.6, 0.9, (100, 100))
axes[0].imshow(healthy, cmap='RdYlGn', vmin=-1, vmax=1)
axes[0].set_title('Healthy Crops (NDVI 0.6-0.9)')
axes[0].colorbar()

# Stressed vegetation
stressed = np.random.uniform(0.1, 0.3, (100, 100))
axes[1].imshow(stressed, cmap='RdYlGn', vmin=-1, vmax=1)
axes[1].set_title('Stressed/Dying (NDVI 0.1-0.3)')
axes[1].colorbar()

# Change map (degradation)
change = np.random.uniform(-0.5, -0.1, (100, 100))
im = axes[2].imshow(change, cmap='RdYlGn', vmin=-1, vmax=1)
axes[2].set_title('NDVI Drop (Crisis Indicator)')
plt.colorbar(im, ax=axes[2])

plt.tight_layout()
plt.show()

# Cell 4: NDVI change calculation demo
from crisiscortex.data.satellite import SentinelDownloader

downloader = SentinelDownloader("dummy", "dummy")

# Simulate before/after crisis
np.random.seed(42)
ndvi_before = np.random.uniform(0.5, 0.8, (50, 50))  # Healthy
ndvi_after = ndvi_before - np.random.uniform(0.2, 0.5, (50, 50))  # Crisis

change = downloader.compute_ndvi_change(ndvi_before, ndvi_after)

fig, axes = plt.subplots(1, 3, figsize=(15, 4))
axes[0].imshow(ndvi_before, cmap='YlGn', vmin=0, vmax=1)
axes[0].set_title('Before Crisis')
axes[1].imshow(ndvi_after, cmap='YlGn', vmin=0, vmax=1)
axes[1].set_title('After Crisis')
im = axes[2].imshow(change, cmap='RdBu_r', vmin=-0.6, vmax=0.6)
axes[2].set_title('NDVI Change (Red = Worse)')
plt.colorbar(im, ax=axes[2])
plt.tight_layout()
plt.show()

# Cell 5: Statistics
print(f"Mean NDVI before: {np.nanmean(ndvi_before):.3f}")
print(f"Mean NDVI after: {np.nanmean(ndvi_after):.3f}")
print(f"Mean change: {np.nanmean(change):.3f}")
print(f"Pixels with >0.3 drop: {np.sum(change < -0.3)} / {change.size}")
```
